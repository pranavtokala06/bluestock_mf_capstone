# Day 3 - Exploratory Data Analysis (EDA)
## Bluestock MF Capstone

Deliverable: EDA_Analysis.ipynb with 15+ charts exported to data/charts/

10 EDA tasks: NAV trends, AUM grouped bar, SIP time-series, category heatmap, demographics, geographic, folio growth, correlation matrix, sector donut, key findings

```python
from pathlib import Path
import pandas as pd, numpy as np, sqlite3
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

sns.set_theme(style='whitegrid', palette='tab10')
plt.rcParams.update({'figure.dpi':130,'axes.spines.top':False,'axes.spines.right':False})

BASE   = Path().resolve().parent
DB     = BASE / "data" / "db" / "bluestock_mf.db"
PROC   = BASE / "data" / "processed"
CHARTS = BASE / "data" / "charts"
CHARTS.mkdir(exist_ok=True)

with sqlite3.connect(DB) as conn:
    nav  = pd.read_sql("SELECT * FROM fact_nav ORDER BY scheme_code, date_key", conn)
    txn  = pd.read_sql("SELECT * FROM fact_transactions", conn)
    perf = pd.read_sql("SELECT p.*, f.scheme_name,f.sub_category,f.category,f.amc_name,f.risk_level FROM fact_performance p JOIN dim_fund f ON p.scheme_code=f.scheme_code", conn)
    aum  = pd.read_sql("SELECT * FROM fact_aum", conn)
    sip  = pd.read_sql("SELECT * FROM ref_sip_inflows ORDER BY month", conn)
    fol  = pd.read_sql("SELECT * FROM ref_folio_count ORDER BY month", conn)
    cat  = pd.read_sql("SELECT * FROM ref_category_inflows", conn)
    ptf  = pd.read_sql("SELECT * FROM ref_portfolio_holdings", conn)
    fund = pd.read_sql("SELECT * FROM dim_fund", conn)
    bm   = pd.read_sql("SELECT * FROM ref_benchmark_indices ORDER BY index_name, date", conn)

nav["date"]  = pd.to_datetime(nav["date_key"])
txn["date"]  = pd.to_datetime(txn["date_key"])
aum["date"]  = pd.to_datetime(aum["date_key"])
sip["month"] = pd.to_datetime(sip["month"])
cat["month"] = pd.to_datetime(cat["month"])
fol["month"] = pd.to_datetime(fol["month"])
bm["date"]   = pd.to_datetime(bm["date"])
print(f"nav={len(nav):,} | txn={len(txn):,} | perf={len(perf)} | aum={len(aum)} | sip={len(sip)} | cat={len(cat)} | fol={len(fol)}")
```

```python
# Chart 1: NAV Trend - All 40 Schemes Indexed to 100
nav_m = nav.merge(fund[['scheme_code','category','sub_category','scheme_name']], on='scheme_code')
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
cats = [('Equity', False), ('Debt', False), ('Hybrid', False), (None, True)]
titles = ['Equity Funds', 'Debt Funds', 'Hybrid Funds', 'Index Funds']

for ax, (cat_name, is_idx), title in zip(axes.flat, cats, titles):
    if is_idx:
        subset = nav_m[nav_m['sub_category'].isin(['Index','Index/ETF'])]
    else:
        subset = nav_m[(nav_m['category']==cat_name) & (~nav_m['sub_category'].isin(['Index','Index/ETF']))]
    for code in subset['scheme_code'].unique():
        fd = subset[subset['scheme_code']==code].sort_values('date')
        if fd.empty: continue
        idx = fd['nav'] / fd['nav'].iloc[0] * 100
        ax.plot(fd['date'], idx, lw=1.2, alpha=0.8, label=fd['scheme_name'].iloc[0][:20])
    ax.axvspan(pd.Timestamp('2023-01-01'), pd.Timestamp('2023-12-31'), alpha=0.07, color='green', label='2023 Bull')
    ax.axvspan(pd.Timestamp('2024-06-01'), pd.Timestamp('2024-10-31'), alpha=0.07, color='red',   label='2024 Correction')
    ax.axhline(100, color='black', lw=0.7, ls='--', alpha=0.4)
    ax.set_title(f'{title} - Indexed NAV (Base=100)', fontweight='bold')
    ax.set_ylabel('Indexed NAV')
    ax.legend(fontsize=5.5, ncol=2, loc='upper left')

plt.suptitle('NAV Trend Analysis - All 40 Schemes (2022-2026)', fontsize=13, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig(CHARTS/'chart01_nav_trends.png', bbox_inches='tight')
plt.show()
print("Chart 1 saved")
```

```python
# Chart 2: AUM Growth Grouped Bar by Fund House
aum['year'] = aum['date'].dt.year
aum_y = aum.groupby(['year','fund_house'])['aum_lakh_crore'].sum().reset_index()
aum_p = aum_y.pivot(index='fund_house', columns='year', values='aum_lakh_crore').fillna(0)

fig, ax = plt.subplots(figsize=(14, 6))
x = np.arange(len(aum_p))
w = 0.2
years = sorted(aum_y['year'].unique())
colors = ['#3498db','#2ecc71','#e67e22','#e74c3c']
for i, (yr, clr) in enumerate(zip(years, colors)):
    if yr in aum_p.columns:
        bars = ax.bar(x + i*w, aum_p[yr], w, label=str(yr), color=clr, alpha=0.85, edgecolor='none')
        for j, (bar, val) in enumerate(zip(bars, aum_p[yr])):
            if aum_p.index[j] == 'SBI Mutual Fund' and val > 0:
                ax.annotate(f'Rs.{val:.1f}L Cr', (bar.get_x()+bar.get_width()/2, val+0.05),
                            ha='center', fontsize=7.5, color='red', fontweight='bold')

ax.set_xticks(x + w*1.5)
ax.set_xticklabels(aum_p.index, rotation=25, ha='right', fontsize=9)
ax.set_ylabel('AUM (Lakh Crore)')
ax.set_title('AUM Growth by Fund House (2022-2025) | SBI MF leads with Rs.12.5L Cr+ (2025)', fontweight='bold')
ax.legend(title='Year')
plt.tight_layout()
plt.savefig(CHARTS/'chart02_aum_by_fundhouse.png', bbox_inches='tight')
plt.show()
print("Chart 2 saved")
```

```python
# Chart 3: SIP Inflow Time-Series with All-Time High Annotation
fig, ax = plt.subplots(figsize=(14, 5))
ax.fill_between(sip['month'], sip['sip_inflow_crore'], alpha=0.25, color='#3498db')
ax.plot(sip['month'], sip['sip_inflow_crore'], color='#3498db', lw=2, marker='o', ms=3)

max_idx = sip['sip_inflow_crore'].idxmax()
max_row = sip.loc[max_idx]
ax.annotate(f"All-Time High\nRs.{max_row['sip_inflow_crore']:,.0f} Cr",
            xy=(max_row['month'], max_row['sip_inflow_crore']),
            xytext=(-80, -45), textcoords='offset points',
            fontsize=9, fontweight='bold', color='green',
            arrowprops=dict(arrowstyle='->', color='green', lw=1.5))

row_20k = sip[sip['sip_inflow_crore'] >= 20000]
if len(row_20k):
    cr20 = row_20k.iloc[0]
    ax.annotate('Crossed Rs.20,000 Cr', xy=(cr20['month'], 20000),
                xytext=(30, 20), textcoords='offset points', fontsize=8, color='darkorange',
                arrowprops=dict(arrowstyle='->', color='darkorange'))

ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'Rs.{x:,.0f}'))
ax.set_title('Monthly SIP Inflows (Jan 2022 - Dec 2025) | Record Rs.31,002 Cr (Dec 2025)', fontweight='bold')
ax.set_ylabel('SIP Inflow (Crore)'); ax.set_xlabel('Month')
plt.tight_layout()
plt.savefig(CHARTS/'chart03_sip_timeseries.png', bbox_inches='tight')
plt.show()
print("Chart 3 saved")
```

```python
# Chart 4: Category Inflow Heatmap
cat['month_label'] = cat['month'].dt.strftime('%b-%y')
pivot = cat.pivot_table(index='category', columns='month_label', values='net_inflow_crore', aggfunc='sum')
month_order = cat.sort_values('month')['month_label'].unique()
pivot = pivot.reindex(columns=month_order)

fig, ax = plt.subplots(figsize=(16, 7))
sns.heatmap(pivot, ax=ax, cmap='RdYlGn', center=0, fmt='.0f',
            annot=True, annot_kws={'size':7.5}, linewidths=0.4,
            cbar_kws={'label':'Net Inflow (Crore)'})
ax.set_title('Category-wise Net Inflows Heatmap (Crore) | Green=inflow, Red=outflow', fontweight='bold', fontsize=12)
ax.set_xlabel('Month'); ax.set_ylabel('Fund Category')
ax.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.savefig(CHARTS/'chart04_category_inflow_heatmap.png', bbox_inches='tight')
plt.show()
print("Chart 4 saved")
```

```python
# Chart 5: Investor Demographics
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
age_counts = txn['age_group'].value_counts().sort_index()
axes[0].pie(age_counts.values, labels=age_counts.index, autopct='%1.1f%%',
            startangle=90, colors=sns.color_palette('Set2', len(age_counts)))
axes[0].set_title('Investor Age Distribution', fontweight='bold')

sip_txn  = txn[txn['transaction_type']=='SIP']
age_order = sorted(sip_txn['age_group'].unique())
bp_data   = [sip_txn[sip_txn['age_group']==ag]['amount_inr'].values for ag in age_order]
bp = axes[1].boxplot(bp_data, labels=age_order, patch_artist=True, showfliers=False,
                      medianprops=dict(color='black', lw=1.5))
for patch, color in zip(bp['boxes'], sns.color_palette('Set2', len(age_order))):
    patch.set_facecolor(color); patch.set_alpha(0.7)
axes[1].set_title('SIP Amount by Age Group', fontweight='bold')
axes[1].set_ylabel('SIP Amount (Rs.)')
axes[1].tick_params(axis='x', rotation=20)

gender_amt = txn.groupby('gender')['amount_inr'].sum()
axes[2].pie(gender_amt.values, labels=gender_amt.index, autopct='%1.1f%%',
            startangle=90, colors=['#3498db','#e74c3c'])
axes[2].set_title('Investment by Gender', fontweight='bold')
plt.suptitle('Investor Demographics Analysis', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig(CHARTS/'chart05_investor_demographics.png', bbox_inches='tight')
plt.show()
print("Chart 5 saved")
```

```python
# Chart 6: Geographic Distribution
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
state_sip = (txn[txn['transaction_type']=='SIP'].groupby('state')['amount_inr'].sum().sort_values() / 1e7)
colors_s  = ['#e74c3c' if v == state_sip.max() else '#3498db' for v in state_sip.values]
axes[0].barh(state_sip.index, state_sip.values, color=colors_s, edgecolor='none')
for i, v in enumerate(state_sip.values):
    axes[0].text(v+0.3, i, f'Rs.{v:.0f}Cr', va='center', fontsize=8)
axes[0].set_xlabel('SIP Amount (Crore)'); axes[0].set_title('SIP Investment by State', fontweight='bold')

tier = txn.groupby('city_tier')['amount_inr'].sum().reset_index()
axes[1].pie(tier['amount_inr'], labels=tier['city_tier'], autopct='%1.1f%%',
            startangle=90, colors=['#2ecc71','#f39c12'], explode=[0.05,0])
axes[1].set_title('T30 vs B30 City Investment Split', fontweight='bold')
plt.suptitle('Geographic Distribution of Investments', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig(CHARTS/'chart06_geographic.png', bbox_inches='tight')
plt.show()
print("Chart 6 saved")
```

```python
# Chart 7: Folio Count Growth with Milestones
fig, ax = plt.subplots(figsize=(13, 5))
ax.plot(fol['month'], fol['total_folios_crore'], 'b-o', lw=2, ms=4, label='Total Folios')
ax.fill_between(fol['month'], fol['equity_folios_crore'], alpha=0.3, color='green', label='Equity Folios')

# Annotate start and end
first = fol.iloc[0]; last = fol.iloc[-1]
ax.annotate(f"{first['total_folios_crore']:.2f} Cr\n({first['month'].strftime('%b %Y')})",
            xy=(first['month'], first['total_folios_crore']),
            xytext=(20,15), textcoords='offset points', fontsize=8.5, color='navy', fontweight='bold',
            arrowprops=dict(arrowstyle='->', color='navy'))
ax.annotate(f"{last['total_folios_crore']:.2f} Cr\n({last['month'].strftime('%b %Y')})",
            xy=(last['month'], last['total_folios_crore']),
            xytext=(-80,15), textcoords='offset points', fontsize=8.5, color='green', fontweight='bold',
            arrowprops=dict(arrowstyle='->', color='green'))

ax.set_title('Industry Folio Count Growth (Crore) | 13.26 Cr to 26.12 Cr - nearly doubled in 3 years',
             fontweight='bold')
ax.set_ylabel('Folios (Crore)'); ax.legend()
plt.tight_layout()
plt.savefig(CHARTS/'chart07_folio_growth.png', bbox_inches='tight')
plt.show()
print("Chart 7 saved")
```

```python
# Chart 8: NAV Return Correlation Matrix - 10 Funds
target_cats = ['Large Cap','Mid Cap','Small Cap','Liquid','Short Duration']
selected_codes = []
for cname in target_cats:
    codes = fund[fund['sub_category']==cname]['scheme_code'].values[:2]
    selected_codes.extend(codes.tolist())
selected_codes = selected_codes[:10]

ret_df = nav[nav['scheme_code'].isin(selected_codes)].copy()
ret_df = ret_df.merge(fund[['scheme_code','scheme_name']], on='scheme_code')
ret_df['short_name'] = ret_df['scheme_name'].str[:18]
pivot  = ret_df.pivot_table(index='date', columns='short_name', values='daily_return')
corr   = pivot.corr()

fig, ax = plt.subplots(figsize=(12, 9))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, ax=ax, mask=mask, cmap='RdYlGn', center=0,
            vmin=-0.2, vmax=1.0, annot=True, fmt='.2f',
            annot_kws={'size':8}, linewidths=0.4)
ax.set_title('Return Correlation Matrix - 10 Selected Funds (Daily Returns 2022-2026)',
             fontweight='bold', fontsize=12)
ax.tick_params(axis='both', labelsize=8)
plt.xticks(rotation=40, ha='right')
plt.tight_layout()
plt.savefig(CHARTS/'chart08_correlation_matrix.png', bbox_inches='tight')
plt.show()
print("Chart 8 saved")
```

```python
# Chart 9: Sector Allocation Donut
eq_codes    = fund[fund['category']=='Equity']['scheme_code'].values
eq_holdings = ptf[ptf['scheme_code'].isin(eq_codes)].copy()
sector_wt   = eq_holdings.groupby('sector')['weight_pct'].mean().sort_values(ascending=False)

threshold = 2.0
main   = sector_wt[sector_wt >= threshold].copy()
others = sector_wt[sector_wt < threshold].sum()
if others > 0:
    main['Others'] = others
main = main.sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(11, 8))
colors_d = sns.color_palette('Set3', len(main))
wedges, texts, autotexts = ax.pie(
    main.values, labels=None, autopct='%1.1f%%',
    startangle=90, pctdistance=0.78, colors=colors_d,
    wedgeprops=dict(width=0.55, edgecolor='white', linewidth=1.5))
for at in autotexts: at.set_fontsize(8)
ax.legend(wedges, main.index, title='Sector', loc='center left',
          bbox_to_anchor=(1,0,0.5,1), fontsize=9)
ax.set_title('Equity Fund Sector Allocation\n(Average weight % across all equity funds)',
             fontweight='bold', fontsize=12)
plt.tight_layout()
plt.savefig(CHARTS/'chart09_sector_allocation.png', bbox_inches='tight')
plt.show()
print("Chart 9 saved")
```

```python
# Chart 10: Returns Distribution + Sharpe Comparison
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
perf_c = perf.dropna(subset=['return_1yr_pct'])
cat_order = perf_c.groupby('sub_category')['return_1yr_pct'].median().sort_values(ascending=False).index
bp_data   = [perf_c[perf_c['sub_category']==c]['return_1yr_pct'].values for c in cat_order]
bp = axes[0].boxplot(bp_data, labels=cat_order, patch_artist=True, showfliers=True,
                      medianprops=dict(color='black', lw=2))
for patch in bp['boxes']:
    patch.set_facecolor('#3498db'); patch.set_alpha(0.65)
axes[0].axhline(0, color='red', lw=1, ls='--', alpha=0.6)
axes[0].set_title('1-Year Return Distribution by Sub-Category', fontweight='bold')
axes[0].set_ylabel('1-Year Return (%)'); axes[0].tick_params(axis='x', rotation=35)

sharpe_c = perf.dropna(subset=['sharpe_ratio']).sort_values('sharpe_ratio', ascending=True)
colors_s = ['#e74c3c' if v < 0 else '#f39c12' if v < 1 else '#27ae60' for v in sharpe_c['sharpe_ratio']]
axes[1].barh([n[:25] for n in sharpe_c['scheme_name']], sharpe_c['sharpe_ratio'],
              color=colors_s, edgecolor='none')
axes[1].axvline(1, color='green', lw=1.2, ls='--', alpha=0.7, label='Sharpe=1')
axes[1].axvline(0, color='black', lw=0.8)
axes[1].set_title('Sharpe Ratio - All 40 Funds\nGreen>1, Amber 0-1, Red<0', fontweight='bold')
axes[1].set_xlabel('Sharpe Ratio'); axes[1].legend(fontsize=8)
axes[1].tick_params(axis='y', labelsize=7.5)
plt.tight_layout()
plt.savefig(CHARTS/'chart10_returns_distribution.png', bbox_inches='tight')
plt.show()
print("Chart 10 saved")
```

```python
# Charts 11-15: Additional Charts
# Chart 11: Avg SIP by age
fig, ax = plt.subplots(figsize=(10, 5))
sip_age = txn[txn['transaction_type']=='SIP'].groupby('age_group')['amount_inr'].mean().sort_values(ascending=False)
bars = ax.bar(sip_age.index, sip_age.values, color=sns.color_palette('husl', len(sip_age)), edgecolor='none')
for bar, val in zip(bars, sip_age.values):
    ax.text(bar.get_x()+bar.get_width()/2, val+50, f'Rs.{val:,.0f}', ha='center', fontsize=9)
ax.set_title('Average SIP Amount by Age Group', fontweight='bold')
ax.set_ylabel('Avg SIP Amount (Rs.)')
plt.tight_layout()
plt.savefig(CHARTS/'chart11_sip_by_age.png', bbox_inches='tight')
plt.show()

# Chart 12: Category AUM pie
fig, ax = plt.subplots(figsize=(9, 7))
cat_aum = perf.groupby('category')['aum_crore'].sum().sort_values(ascending=False)
ax.pie(cat_aum.values, labels=cat_aum.index, autopct='%1.1f%%',
       startangle=90, colors=sns.color_palette('Set2', len(cat_aum)))
ax.set_title('AUM Distribution by Category', fontweight='bold')
plt.tight_layout()
plt.savefig(CHARTS/'chart12_aum_by_category.png', bbox_inches='tight')
plt.show()

# Chart 13: Folio stacked area
fig, ax = plt.subplots(figsize=(13, 5))
ax.stackplot(fol['month'],
             fol['equity_folios_crore'], fol['debt_folios_crore'],
             fol['hybrid_folios_crore'], fol['others_folios_crore'],
             labels=['Equity','Debt','Hybrid','Others'],
             colors=['#27ae60','#3498db','#e67e22','#95a5a6'], alpha=0.8)
ax.set_title('Folio Count by Category - Stacked Area', fontweight='bold')
ax.set_ylabel('Folios (Crore)'); ax.legend(loc='upper left')
plt.tight_layout()
plt.savefig(CHARTS/'chart13_folio_stacked.png', bbox_inches='tight')
plt.show()

# Chart 14: T30 vs B30 by type
fig, ax = plt.subplots(figsize=(10, 5))
tier_type = txn.groupby(['city_tier','transaction_type'])['amount_inr'].sum().unstack().fillna(0) / 1e7
tier_type.plot(kind='bar', ax=ax, color=['#e74c3c','#f39c12','#27ae60'], edgecolor='none')
ax.set_title('Transaction Amount by City Tier and Type (Crore)', fontweight='bold')
ax.set_ylabel('Amount (Crore)'); ax.tick_params(axis='x', rotation=0)
plt.tight_layout()
plt.savefig(CHARTS/'chart14_t30_b30_types.png', bbox_inches='tight')
plt.show()

# Chart 15: Volatility heatmap (monthly)
nav_m2 = nav.merge(fund[['scheme_code','scheme_name']], on='scheme_code')
nav_m2['ym'] = nav_m2['date'].dt.to_period('M')
monthly_vol  = nav_m2.groupby(['scheme_name','ym'])['daily_return'].std() * np.sqrt(21) * 100
pivot_vol    = monthly_vol.unstack()
pivot_vol    = pivot_vol.iloc[:, -12:] if pivot_vol.shape[1] >= 12 else pivot_vol
top_v        = pivot_vol.mean(axis=1).nlargest(10).index
pv_top       = pivot_vol.loc[top_v]
pv_top.index = [n[:22] for n in pv_top.index]
pv_top.columns = [str(c) for c in pv_top.columns]

fig, ax = plt.subplots(figsize=(14, 6))
sns.heatmap(pv_top, ax=ax, cmap='YlOrRd', fmt='.1f', annot=True,
            annot_kws={'size':7}, linewidths=0.3, cbar_kws={'label':'Monthly Vol %'})
ax.set_title('Monthly Return Volatility Heatmap (%) - Top 10 Volatile Funds', fontweight='bold')
ax.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.savefig(CHARTS/'chart15_volatility_heatmap.png', bbox_inches='tight')
plt.show()
print("Charts 11-15 saved")
```

## 10 Key EDA Findings

1. **Equity dominance:** Equity funds hold ~65% of industry folios (16.9 Cr of 26.1 Cr), confirming retail preference for wealth creation.

2. **SIP structural growth:** Monthly SIP grew 2.7x from ~Rs.11,500 Cr (Jan 2022) to Rs.31,002 Cr (Dec 2025), driven by UPI/NACH adoption.

3. **SBI AUM dominance:** SBI MF commands ~18% market share at Rs.12.5+ Lakh Crore in 2025, consistently outpacing peers since 2022.

4. **Geographic concentration:** Maharashtra, Karnataka, and Delhi NCR account for >50% of SIP inflows; B30 cities contribute ~22% - significant room for growth.

5. **Inter-fund correlations 0.4-0.75:** Small-cap vs debt correlation near 0, confirming multi-asset diversification value (Chart 8).

6. **Financial Services sector dominates:** Avg 28% weight in equity funds - highest single-sector concentration risk (Chart 9).

7. **Millennials lead SIP:** 26-35 age group has highest SIP transaction count with avg ticket ~Rs.7,500 (Chart 5).

8. **Debt category net outflows:** Category heatmap shows consistent outflows from Money Market/Liquid as investors shift to equity (Chart 4).

9. **Mid/Small-cap 2024 correction:** Monthly volatility heatmap shows volatility spike Jun-Sep 2024 aligned with FII selloffs (Chart 15).

10. **Folio count doubled:** From 13.26 Cr to 26.12 Cr in 3 years - strongest growth in equity folios, SIP-first investors entering market rapidly.
